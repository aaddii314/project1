#### Survey data on smoking habits from the United Kingdom. The data set can be used for analyzing the demographic characteristics of smokers and types of tobacco consumed. A data frame with 1691 observations on the following 12 variables.


>**gender** - Gender with levels Female and Male.
>
>**age** - Age.
>
>**marital_status** - Marital status with levels Divorced, Married, Separated, Single and Widowed.
>
>**highest_qualification** - Highest education level with levels A Levels, Degree, GCSE/CSE, GCSE/O Level,
>Higher/Sub Degree, No Qualification, ONC/BTEC and Other/Sub Degree
>
>**nationality** - Nationality with levels British, English, Irish, Scottish, Welsh, Other, Refused and Unknown.
>
>**ethnicity** - Ethnicity with levels Asian, Black, Chinese, Mixed, White and Refused Unknown.
>
>**gross_income** - Gross income with levels Under 2,600, 2,600 to 5,200, 5,200 to 10,400, 10,400 to 15,600, 15,600 to 20,800, 20,800 to 28,600, 28,600 to 36,400, Above 36,400, Refused and Unknown.
>
>**region** - Region with levels London, Midlands And East Anglia, Scotland, South East, South West, The North and Wales
>
>**smoke** - Smoking status with levels No and Yes
>
>**amt_weekends** - Number of cigarettes smoked per day on weekends.
>
>**amt_weekdays** - Number of cigarettes smoked per day on weekdays.
>
>**type** - Type of cigarettes smoked with levels Packets, Hand-Rolled, Both/Mainly Packets and Both/Mainly Hand-Rolled

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/aaddii314/project1/main/smoking.csv'
smoke_df = pd.read_csv(url)
```

```python
smoke_df.shape
```

```python
smoke_df.columns
```

```python
smoke_df.head()
```

#### <span style="color:blue"> Note:</span>
We've grouped the types of degrees and certificates to provide a broader overview and clearer insights into the differences between education levels. This rough categorization aims to offer a more straightforward understanding of the disparities across various educational backgrounds.

```python
smoke_df["highest_qualification"]=smoke_df["highest_qualification"].map({"GCSE/CSE": "A high school diploma","GCSE/O Level":"A high school diploma","A Levels":"A high school diploma","Higher/Sub Degree":"Diploma","ONC/BTEC":"Diploma","Other/Sub Degree":"Diploma","No Qualification":"No Qualification","Degree":"Degree"})
```

#### <span style="color:blue"> Note:</span>
We defined the "gross-income" into tiers 1-8.

```python
smoke_df["gross_income"]=smoke_df["gross_income"].map({"Under 2,600":1,"2,600 to 5,200":2, "5,200 to 10,400":3,"10,400 to 15,600":4,"15,600 to 20,800":5,"20,800 to 28,600":6,"28,600 to 36,400":7,"Above 36,400":8})
```

#### <span style="color:blue"> Note:</span>
We have grouped the types of cigarettes into "Packets" and "Hand-rolled".

```python
smoke_df["type"]=smoke_df["type"].map({"Both/Mainly Packets":"Packets","Both/Mainly Hand-Rolled":"Hand-Rolled","Packets":"Packets","Hand-Rolled":"Hand-Rolled"})
```

```python
smoke_df=smoke_df.iloc[:,1:]
smoke_df.shape
```

```python
smoke_df
```

#### <span style="color:blue"> Note:</span>
We added a column of "avg" which represents the average cigarettes smoked per day (both weekends and weekdays) which we will use later.

```python
smoke_df["avg"]=(smoke_df["amt_weekends"]+smoke_df["amt_weekdays"])/2
```

```python
smoke_df.groupby("gender")["smoke"].count()
```

```python
smoking_female=smoke_df[(smoke_df.gender=="Female")]
```

```python
smoking_male=smoke_df[(smoke_df.gender=="Male")]
```

```python
fig, axes = plt.subplots(figsize=(7, 3), ncols=2)
sns.histplot(data=smoking_male, x='smoke',stat='percent', ax = axes[0])
sns.histplot(data=smoking_female, x='smoke',stat='percent', ax = axes[1])
axes[0].set_xlabel("male")
axes[1].set_xlabel("female")
plt.show()
```

## **Here we can see that men and women smoke equally, both around 25%.**

```python
fig, axes = plt.subplots (1,2 ,figsize=(5, 4))
plt.subplots_adjust(wspace = 0.5)
sns.boxplot(data=smoking_male, y='avg', ax = axes[0])
sns.boxplot(data=smoking_female, y='avg', ax = axes[1])
axes[0].set_ylim([0, 40])
axes[1].set_ylim([0, 40])
axes[0].set_xlabel("male")
axes[1].set_xlabel("female")
```

### **We observe that the median smoking rates for men and women are similar, with men smoking an average of 15 cigarettes per day and women smoking an average of 13.5 cigarettes per day. This represents a difference of approximately 10% between the two groups.**

```python
fig, axes = plt.subplots (1,2 ,figsize=(5, 5))
plt.subplots_adjust(wspace = 0.5)
sns.boxplot(data=smoking_male, y='amt_weekdays', ax = axes[0])
sns.boxplot(data=smoking_female, y='amt_weekdays', ax = axes[1])
axes[0].set_ylim([0, 40])
axes[1].set_ylim([0, 40])
axes[0].set_xlabel("male")
axes[1].set_xlabel("female")
```

## **On the other hand, if we consider only weekdays, a more pronounced difference emerges, with men smoking an average of 15 cigarettes per day compared to women, who smoke around 10 cigarettes per day. This represents a significant difference of 50% between the two groups.**

```python
sns.displot(data=smoke_df, x='age', hue='smoke', fill=True, height=5, aspect=1.5,bins=8)
plt.title('age vs smoke')
```

### **The graph illustrates a notable decline in both the number and percentage of smokers as age increases.**

```python
sns.lmplot(x='age', y='avg', data=smoke_df, scatter=False)

# Add title and labels
plt.title('Linear Relationship between age and avg')
plt.xlabel('Age')
plt.ylabel('avg')

# Show the plot
plt.show()
```

### **As age increases, the average number of cigarettes per day for a smoker increases.**

# <span style="color:green"> Observation 2:</span>
### **It seems there are two different approaches to quitting smoking: immediate cessation or gradual reduction. In Britain, immediate cessation is commonly favored, as evidenced by the decreasing number of smokers with age, despite an increase in the average cigarettes smoked per day.**
> ### **We recommend making the gradual reduction method more accessible for those who struggle to quit with the immediate cessation approach.**

```python
smoke_young=smoke_df[(smoke_df.age<30)]
smoke_adult=smoke_df[(smoke_df.age>30)]
```

```python
fig, axes = plt.subplots(figsize=(5, 5), ncols=2)
plt.subplots_adjust(wspace = 0.5)

sns.boxplot(data=smoke_adult, y='amt_weekdays', ax = axes[0])
sns.boxplot(data=smoke_adult, y='amt_weekends', ax = axes[1])
axes[0].set_ylim([0, 40])
axes[1].set_ylim([0, 40])
axes[0].set_xlabel("amount_weekdays")
axes[1].set_xlabel("amount_weekends")
axes[0].set_ylabel("")
axes[1].set_ylabel("")
axes[0].set_title("adult")
axes[1].set_title("adult")
```

### **There is no substantial increase in smoking among adults from the weekend to the middle of the week.**

```python
fig, axes = plt.subplots(figsize=(5, 5), ncols=2)
plt.subplots_adjust(wspace = 0.5)

sns.boxplot(data=smoke_young, y='amt_weekdays', ax = axes[0])
sns.boxplot(data=smoke_young, y='amt_weekends', ax = axes[1])
axes[0].set_ylim([0, 40])
axes[1].set_ylim([0, 40])

axes[0].set_xlabel("amount_weekdays")
axes[1].set_xlabel("amount_weekends")
axes[0].set_ylabel("")
axes[1].set_ylabel("")
axes[0].set_title("young")
axes[1].set_title("young")
```

### **We see a big increase in smoking among young people under 30 smoking on the weekend compared to the middle of the week**
